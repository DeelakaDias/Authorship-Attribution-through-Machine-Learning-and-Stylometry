```python
user1_txt = open('./datasets/chats/user1.txt', 'r')
user1_chats = user1_txt.read().split('\n')
user1_txt.close()
```

```python
def remove_non_ascii(text):
    return ''.join(i for i in text if ord(i)<128)
```

```python
for i in range(len(user1_chats)):
    user1_chats[i] = remove_non_ascii(user1_chats[i])

user1_chats = [x for x in user1_chats if x != '']
```

```python
print(user1_chats)
```

# Randomize the chat order to preserve privacy

Shuffle all messages in the chat so the chat looks meaningless. This increases the privacy of people who provide their chat history to us for research purpose.

```python
# Fisher Yates Shuffle to randomize the order of the chats
import random

def shuffle(array):
    for i in range(len(array)):
        j = random.randint(0, i)
        array[i], array[j] = array[j], array[i]

shuffle(user1_chats)
```

# Generate yml file to be used in few shot prompting and its testing

Fields of yml file such as word language is not generated by this script. Those labels should be manually added.

```python
yml_representation = ""

for i in range(len(user1_chats)):
    sentence = user1_chats[i]
    yml_representation += "- sentence: " + sentence + "\n"
    yml_representation += "  words: " + "\n"
    
    words = sentence.split()
    for j in range(len(words)):
        yml_representation += "    - word: " + words[j] + "\n"
        yml_representation += "      lang: " + "\n"
    
    yml_representation += "\n"
```

```python
with open('./singlish-mix-text/sample1.yml', 'w') as f:
    f.write(yml_representation)
```
